```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import sqlite3 as sql
```

```python
name=[]
price=[]
rating=[]
battery=[]
ram=[]
display=[]
camera=[]
```

```python
pages=list(range(1,25))
for page in pages:
    res=requests.get("https://www.flipkart.com/search?q=redmi+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_3_7_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_3_7_na_na_ps&as-pos=3&as-type=HISTORY&suggestionId=redmi+mobile%7CMobiles&requestId=c0531f54-c240-4089-a9dd-a0496edc69a5&as-backfill=on{}".format(page))
    st=res.text
    bs=BeautifulSoup(st,"html.parser")
    names=bs.find_all("div",{"class":"_4rR01T"})
    for x in range(len(names)):
        name.append(names[x].text)
    prices=bs.find_all("div",{"class":"_30jeq3 _1_WHN1"})
    for x in range(len(prices)):
        price.append(prices[x].text)
    ratings=bs.find_all("span",{"class":"_2_R_DZ"})
    for x in range(len(ratings)):
        rating.append(ratings[x].text)
    features=bs.find_all("li",{"class":"rgWa7D"})
    for x in range(len(features)):
        d=features[x].text
        if "Display" in d:
            display.append(d)
        elif "Battery" in d:
            battery.append(d)
        elif "RAM" in d:
            ram.append(d)
            
        elif "Camera" in d:
            camera.append(d)
        
    
```

```python
print(len(name))
print(len(price))
print(len(rating))
print(len(display))
print(len(battery))
print(len(ram))
print(len(camera))
```
