```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
description=[]
cos=[]
processor=[]
ram=[]
storage=[]
os=[]
display=[]
warrenty=[]
rate=[]
pages=list(range(1,31))
for page in pages:
    res = requests.get(
        "https://www.flipkart.com/search?q=laptop&otracker=AS_Query_HistoryAutoSuggest_2_0&otracker1=AS_Query_HistoryAutoSuggest_2_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=2&as-type=HISTORY&as-backfill=on{}".format(page))
    st = res.text
    bs = BeautifulSoup(st, "html.parser")
    desc = bs.findAll("div", {"class": "_4rR01T"})
    for x in range(len(desc)):
        description.append(desc[x].text)
    common = bs.find_all("li", class_="rgWa7D")
    for x in range(0, len(common)):
        p = common[x].text
        if ("Processor" in p):
            processor.append(p)
        elif ("RAM" in p):
            ram.append(p)
        elif "Display" in p:
            display.append(p)
        elif "Warranty" in p:
            warrenty.append(p)

        elif ("SSD" in p or "HDD" in p):
            storage.append(p)

    price = bs.findAll("div", {"class": "_30jeq3 _1_WHN1"})
    for x in range(len(price)):
        cos.append(price[x].text)
    rating = bs.findAll("div", {"class": "gUuXy-"})
    for x in range(len(rating)):
        rate.append(rating[x].text)
print("des",len(description))
print("pro",len(processor))
print(len(display))
print(len(warrenty))
print(len(storage))

print(len(ram))
print(len(cos))
print(len(rate))
```

```python
df={"desc":description[slice(660)],"cost":cos[slice(660)],"rateing":rate[slice(660)],"processor":processor[slice(660)],"display":display[slice(660)],"warrenty":warrenty[slice(660)],"storage":storage[slice(660)],"ram":ram[slice(660)]}
dataset=pd.DataFrame(data=df)
print(dataset)
```
